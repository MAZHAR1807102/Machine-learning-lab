```python
def loss_bce(y_true, y_pred, W):
    # Binary cross entropy loss
    #loss = (-1)/y_true.shape[0] * ( np.sum( (y_true * np.log(y_pred)) + ((1-y_true) * np.log(1-y_pred))) )
    loss = (-1)/100 * ( np.sum( (y_true * np.log(y_pred)) + ((1-y_true) * np.log(1-y_pred))) )
    
    return loss
```
